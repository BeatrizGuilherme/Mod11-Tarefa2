```python
#Importando as bibliotecas necessárias para o exercício.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
#Carregando o Dataframe
boston = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
boston.head()
```

```python
#Criando um Dataframe sem a variável resposta.
X= boston.drop(columns=['medv']).copy()
X.head()
```

```python
#Criando um Dataframe só com a variável resposta. 
y= boston[['medv']]
y.head()
```

### 1. Execute os passos do exercício anterior, até que você tenha uma árvore de regressão predizendo o valor do imóvel na base de treinamento.

```python
#Separando os dados para validação de teste
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2360873)
```

```python
#Treinando uma árvore com profundidade máxima 2.
arv_1 = DecisionTreeRegressor(max_depth = 2)
arv_1.fit(X_train, y_train)
```

```python
#Cálculo de MSE da árvore com profundidade máxima 2.
mse1= arv_1.score(X_train, y_train)
mse1
```

```python
#Plotando a árvore com profundidade máxima 2.

plt.rc ('figure', figsize = (10,10))
tp = tree.plot_tree(arv_1,feature_names=X.columns,filled= True) 
```

### 2.  Calcule o caminho indicado pelos CCP-alfas dessa árvore.

```python
path = arv_1.cost_complexity_pruning_path(X_train, y_train)
path
```

```python
ccp_alphas, impurities = path.ccp_alphas, path.impurities

plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, impurities)
plt.xlabel("Alpha efetivo")
plt.ylabel("Impureza total das folhas")
```

### 3. Paca cada valor de alpha obtido no item 2, treine uma árvore com o respectivo alfa, e guarde essa árvore em uma lista.

```python
clfs = []

for ccp_alpha in ccp_alphas:
    clf = DecisionTreeRegressor(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)
```

```python
tree_depths = [clf.tree_.max_depth for clf in clfs]
plt.figure(figsize=(10,  6))
plt.plot(ccp_alphas[:-1], tree_depths[:-1])
plt.xlabel("effective alpha")
plt.ylabel("Profundidade da árvore")
```

### 4. Para cada árvore na lista, calcule o MSE da árvore.

```python
arv_1 = DecisionTreeRegressor(max_depth=3, min_samples_leaf=10)

arv_1.fit(X_train, y_train)

mse1 = arv_1.score(X_test, y_test)
mse1
```

```python
mses = []
ind_i = []
ind_j = []

for i in range(2, 9):
    for j in range(2, 10):
        arv_1 = DecisionTreeRegressor(max_depth=i, min_samples_leaf=j)
        arv_1.fit(X_train, y_train)#mandando treinar a árvore
        mse1 = arv_1.score(X_test, y_test)#Calculando o R2 na base de teste
        mses.append(mse1)
        ind_i.append(i)
        ind_j.append(j)
        
df_mse = pd.DataFrame({'mses':mses, 'profundidade':ind_i, 'n_minimo':ind_j})
sns.heatmap(df_mse.pivot(index='profundidade', columns='n_minimo', values='mses'))
df_mse.pivot(index='profundidade', columns='n_minimo', values='mses')
```

### 5. Monte um gráfico do MSE pelo alpha, escolha um valor de alpha perto do ponto de mínimo do MSE

```python

train_scores = [mean_squared_error(y_train , clf.predict(X_train)) for clf in clfs]
test_scores = [mean_squared_error(y_test, clf.predict(X_test)) for clf in clfs]

fig, ax = plt.subplots()
ax.set_xlabel('alpha')
ax.set_ylabel('MSE')
ax.set_title('MSE x alpha do conjunto de dados de treino e teste')
ax.plot(ccp_alphas[:-1], train_scores [:-1], marker = 'o', label = 'treino', drawstyle = 'steps-post')
ax.plot(ccp_alphas[:-1], test_scores [:-1], marker = 'o', label = 'teste', drawstyle = 'steps-post')
ax.legend()
plt.show()
```
